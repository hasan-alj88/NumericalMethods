# Q1
 $$\min f(x) = x^2 + 10 \sin(x), x \in [-10,10] $$

```python
import numpy as np
import sympy as sp
from IPython.core.display import Markdown
from seaborn.external.docscrape import header

from Optimazation.GAExtrema import GAExtrema
from utils.LaTeXTools import df_to_latex
from pathlib import Path

tables_path = Path().cwd() / 'latex' / 'tables'
tables_path.mkdir(exist_ok=True)
```

```python


x_sym = sp.Symbol('x')

q1_data = dict(
    function = lambda x: x**2 + 10*np.sin(x),
    function_sympy = x_sym**2 + 10*sp.sin(x_sym),
    x_min = -10,
    x_max = 10,
)
```

## Gradient Descent

```python
from Optimazation.GradientDescent import GradientDescent
q1_a_solver = GradientDescent(
    function=q1_data.get('function_sympy'),
    learning_rate=0.1,
    initial_guess=dict(x=0)
)
q1_a_table = q1_a_solver.run()
```

```python
display(Markdown(f"**Gradient Descent**"))
display(Markdown(q1_a_table.to_markdown()))
```

```python
# save table
df_to_latex(
    df=q1_a_table,
    filepath=tables_path / 'q1_Gradient_decent.tex',
    caption='Gradient Descent for f(x) = x^2 + 10 \sin(x)',
    formatting=dict(
        x=dict(header='$x$'),
        Gradient_x=dict(header=r'$\frac{df}{dx}$')
    )
)
```

## Golden Search

```python
from Optimazation.GoldenSectionSearch import GoldenSectionSearch

q1_b_solver = GoldenSectionSearch(
    function=q1_data.get('function'),
    x_lower=q1_data.get('x_min'),
    x_upper=q1_data.get('x_max'),
)
q1_b_table = q1_b_solver.run()
```

```python
display(Markdown(f"**Golden Section Search**"))
display(Markdown(q1_b_table.to_markdown()))
```

```python
# save table
df_to_latex(
    df=q1_b_table,
    filepath=tables_path / 'q1_b_GoldenSearch.tex',
    caption='Golden Section Search for f(x) = x^2 + 10 \sin(x)'
)
```

# Genetic algorithm

```python
q1_c_solver = GAExtrema(
    function=q1_data.get('function'),
    x_lower=q1_data.get('x_min'),
    x_upper=q1_data.get('x_max'),
    find_max=False,
    num_generations=30,
    num_elites=2,
    num_parents_mating=6,
    mutation_probability=0.1,
)

q1_c_table = q1_c_solver.run()
```

```python
display(Markdown(f"**Genetic Algorithm**"))
q1_c_table = q1_c_table.loc[:, list(map(str,range(len(q1_c_table.columns)-1)))+['f(x)']]
display(Markdown(q1_c_table.to_markdown()))
```

```python
# save table
df_to_latex(
    df=q1_c_table,
    filepath=tables_path / 'q1_c_GA.tex',
    caption='Genetic Algorithm for f(x) = x^2 + 10 \sin(x)'
)
```

# Q2

$$ \min f(x,y) = x^2+y^2+10 \sin(x+y) , x,y \in [-5,5]  $$

```python
x_sym = sp.Symbol('x')
y_sym = sp.Symbol('y')
q2_data = dict(
    function_sympy = x_sym**2 + y_sym**2 + 10*sp.sin(x_sym+y_sym),
    x_min = -5,
    x_max = 5,
    y_min = -5,
    y_max = 5,
)
```

```python
q2_a_solver = GradientDescent(
    function=q2_data.get('function_sympy'),
    learning_rate=0.05,
    initial_guess=dict(x=0,y=0),
    max_iterations=100,
)
q2_a_table = q2_a_solver.run()
```

```python
display(Markdown(f"**Gradient Descent**"))
display(Markdown(q2_a_table.to_markdown()))
```

```python
# save table
df_to_latex(
    df=q2_a_table,
    filepath=tables_path / 'q2_Gradient_decent.tex',
    caption='Gradient Descent for f(x,y) = x^2+y^2+10 \sin(x+y)',
    formatting=dict(
        x=dict(header='$x$'),
        Gradient_x=dict(header=r'$\frac{df}{dx}$'),
        Gradient_y=dict(header=r'$\frac{df}{dy}$')
    )
)
```

# Q3

Minmize Rastrign function

$$ f(x,y) = 20 + x^2 - 10 cos(2\pi x)+y^2 - 10 cos(2\pi y) , x,y \in[-5.12, 5.12]$$

```python
q3_data = dict(
    function_sympy = 20 + x_sym**2 - 10*sp.cos(2*sp.pi*x_sym) + y_sym**2 - 10*sp.cos(2*sp.pi*y_sym),
    x_min = -5.12,
    x_max = 5.12,
    y_min = -5.12,
    y_max = 5.12,
)
```

```python
q3_a_solver = GradientDescent(
    function=q3_data.get('function_sympy'),
    learning_rate=0.01,
    initial_guess=dict(x=2.5,y=2.5),
    max_iterations=30,
)
q3_a_table_1 = q3_a_solver.run()
```

```python
display(Markdown(f"**Gradient Descent**"))
display(Markdown(q3_a_table_1.to_markdown()))
```
